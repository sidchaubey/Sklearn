# Improving Model Accuracy of Linear Regression
- Linear Regression is a Supervised type of machine learning which tries to fit a line such that it can be used to make predictions. As the dimension of model increases this linear model fails to achieve significance level of accuracy. 
- A hack to improve the accuracy of Linear Regression is by adding more number of columns by multiplying one column by other.
 - Example suppose we have equation like **y= m1x1+m2x2+c** we can add a new dimension by changing equation as : **y = m1x1+m2x2+m3(x1^2)+m4(x2^2)+m5(x1*x2)+c**

```
from sklearn import datasets
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```
boston = datasets.load_boston()
# boston
```

```
X = pd.DataFrame(data=boston.data, columns=boston.feature_names)
X
```

```
X2 = X.copy()
for i in X2.columns:
  X2[i+"_"+i] = X[i]**2
```

```
display(X.head())
display(X2.head())
```

```
Y = pd.DataFrame(data= boston.target, columns=["result"])
Y2 = Y.copy()
```

```
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,Y, random_state=0)
X2_train, X2_test, Y2_train, Y2_test = model_selection.train_test_split(X2,Y2, random_state=0)
```

```
# Fitting for X
lr = LinearRegression()
lr.fit(X_train, Y_train)
Y_pred = lr.predict(X_test)
print(lr.score(X_test, Y_test))
```

```
# Fitting for X
lr2 = LinearRegression()
lr2.fit(X2_train, Y2_train)
Y2_pred = lr2.predict(X2_test)
print(lr2.score(X2_test, Y2_test))
```

```
print(len(lr.coef_[0]))
print(len(lr2.coef_[0]))
```

```
print(lr.intercept_)
print(lr2.intercept_)
```
